```python
def get_outlier_prop(x):
    Q1, Q3 = x.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    outliers = x[(x > upper_bound) | (x < lower_bound)]

    return str(round(100 * len(outliers) / len(x), 1)) + '%'

data.apply(get_outlier_prop)
```
